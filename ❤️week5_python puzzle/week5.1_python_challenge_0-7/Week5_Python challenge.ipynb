```python
#Python Challenge 1
#link: http://www.pythonchallenge.com/pc/def/map.html
def plustwo(letter):
  if letter.isalpha()==True:
      if letter=="y":
        return "a"
      elif letter=="z":
        return "b"
      else:
        return chr(ord(letter)+2)
  else:
    return letter

def trans(mystr):
 res=""
 for x in mystr:
   res+=plustwo(x)
 return res
trans('abc')
```

```python
from urllib import request
from PIL import Image

url = 'http://www.pythonchallenge.com/pc/def/oxygen.png'
response = request.urlopen(url)
oxygen = Image.open(response)
oW, oH = oxygen.size
h = oH // 2
msg = ''
for w in range(0, oW, 7):
    pixel = oxygen.getpixel((w, h))
    msg  += chr(pixel[0])

# print(msg)

s = msg.index('[')
e = msg.index(']')
code = msg[s+1:e].split(',')
for c in code:
    print(chr(int(c)), end='')

print('')
```
